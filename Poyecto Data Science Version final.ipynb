### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Ingreso_tienda1 = tienda['Precio'].sum()
Ingreso_tienda2 = tienda2['Precio'].sum()
Ingreso_tienda3 = tienda3['Precio'].sum()
Ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
precio_total_tienda1 = float(Ingreso_tienda1)
precio_total_tienda2 = float(Ingreso_tienda2)
precio_total_tienda3 = float(Ingreso_tienda3)
precio_total_tienda4 = float(Ingreso_tienda4)
```

# 2. Ventas por categoría

```python
venta_categoria1 = tienda.groupby(['Categoría del Producto']).size().reset_index(name='Cantidad')
venta_categoria2 = tienda2.groupby(['Categoría del Producto']).size().reset_index(name='Cantidad')
venta_categoria3 = tienda3.groupby(['Categoría del Producto']).size().reset_index(name='Cantidad')
venta_categoria4 = tienda4.groupby(['Categoría del Producto']).size().reset_index(name='Cantidad')
```

# 3. Calificación promedio de la tienda


```python
calif_prom_tienda1 = round(float(tienda['Calificación'].mean()))
calif_prom_tienda2 = round(float(tienda2['Calificación'].mean()))
calif_prom_tienda3 = round(float(tienda3['Calificación'].mean()))
calif_prom_tienda4 = round(float(tienda4['Calificación'].mean()))
```

```python
calif_prom_tienda1
```

# 4. Productos más y menos vendidos

```python
mas_vendido1 = tienda.groupby(['Producto']).size().reset_index(name='Cantidad').sort_values(by='Cantidad', ascending=False)
mas_vendido2 = tienda2.groupby(['Producto']).size().reset_index(name='Cantidad').sort_values(by='Cantidad', ascending=False)
mas_vendido3 = tienda3.groupby(['Producto']).size().reset_index(name='Cantidad').sort_values(by='Cantidad', ascending=False)
mas_vendido4 = tienda4.groupby(['Producto']).size().reset_index(name='Cantidad').sort_values(by='Cantidad', ascending=False)
```

```python
mas_vendido1
```

# 5. Envío promedio por tienda

```python
envio_promedio1 = round(float(tienda['Costo de envío'].mean()),2)
envio_promedio2 = round(float(tienda2['Costo de envío'].mean()),2)
envio_promedio3 = round(float(tienda3['Costo de envío'].mean()),2)
envio_promedio4 = round(float(tienda4['Costo de envío'].mean()),2)
```

```python
envio_promedio1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Gráfico de barras: Ingresos por tienda
ingresos = [precio_total_tienda1, precio_total_tienda2, precio_total_tienda3, precio_total_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 6))
sns.barplot(x=tiendas, y=ingresos, palette="viridis")
plt.title("Ingresos por tienda")
plt.ylabel("Ingreso total ($)")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()

# 2. Gráfico de pastel: Distribución por categoría de productos (Tienda 1)
plt.figure(figsize=(8, 6))
plt.pie(venta_categoria1['Cantidad'], labels=venta_categoria1['Categoría del Producto'], autopct='%1.1f%%', startangle=140)
plt.title("Distribución de productos por categoría (Tienda 1)")
plt.axis('equal')  # Para que sea circular
plt.tight_layout()
plt.show()

# 3. Gráfico de líneas: Calificación promedio por tienda
calificaciones = [calif_prom_tienda1, calif_prom_tienda2, calif_prom_tienda3, calif_prom_tienda4]

plt.figure(figsize=(8, 6))
sns.lineplot(x=tiendas, y=calificaciones, marker='o', linewidth=2)
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación promedio")
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```

Informe Final: Análisis Comparativo de Tiendas para Decisión Comercial del Sr. Juan
Introducción
El objetivo de este análisis es asistir al Sr. Juan en la toma de decisiones sobre en qué tienda debe vender sus productos. Para ello, se han analizado los datos de cuatro tiendas diferentes, evaluando variables clave como ingresos totales, distribución por categoría, calificaciones de los clientes, productos más vendidos y costos promedio de envío. A través de visualizaciones y métricas, se identificaron patrones relevantes que orientan una recomendación fundamentada.

Desarrollo
1. Ingresos Totales por Tienda
Los ingresos acumulados por tienda muestran que:

Tienda 1: $25,416.00

Tienda 2: $24,576.00

Tienda 3: $23,648.00

Tienda 4: $24,360.00

La Tienda 1 lidera con el mayor ingreso total, lo que sugiere un buen volumen de ventas y/o precios competitivos. Esto se refleja en el gráfico de barras, donde destaca ligeramente frente a las demás.

2. Distribución por Categoría de Productos
A través de gráficos de pastel y tablas, se observa que:

En la Tienda 1, la categoría más vendida es Bebidas, seguida de Snacks.

Las categorías menos vendidas varían, pero productos como Dulces y Lácteos suelen tener menor participación.

Este patrón puede indicar una oportunidad para entrar en una tienda donde su producto pertenezca a una categoría fuerte y tenga alta rotación.

3. Calificación Promedio de Clientes
Las calificaciones promedio por tienda son:

Tienda 1: 4.0

Tienda 2: 4.0

Tienda 3: 3.0

Tienda 4: 3.0

Las Tiendas 1 y 2 muestran mejor percepción del cliente, lo cual puede influir directamente en la fidelización y repetición de compra.

4. Productos Más y Menos Vendidos
El análisis de frecuencia de productos revela que:

Las Tiendas 1 y 2 tienen mayor concentración de ventas en un grupo reducido de productos, lo cual puede ser estratégico si el Sr. Juan logra posicionar un producto en ese grupo.

Las Tiendas 3 y 4 tienen una venta más distribuida pero con menor volumen total.

5. Costo de Envío Promedio
Los costos de envío promedio son:

Tienda 1: $8.56

Tienda 2: $8.65

Tienda 3: $8.61

Tienda 4: $8.57

Aunque los valores son similares, la Tienda 1 tiene el menor costo promedio, lo cual puede ser una ventaja tanto para el cliente como para el vendedor en términos de competitividad.

Conclusión y Recomendación
Después de evaluar todos los factores clave, se recomienda que el Sr. Juan venda sus productos en la Tienda 1. Esta tienda se destaca por:

Tener los mayores ingresos totales.

Contar con una alta calificación de los clientes.

Poseer una categoría dominante y bien definida en la que podría posicionar su producto.

Tener el menor costo promedio de envío, lo cual mejora la experiencia de compra.

Mostrar un patrón de productos más vendidos, lo que sugiere un entorno favorable para ventas concentradas.

Aunque la Tienda 2 también presenta buenos resultados, la ligera ventaja en ingresos, costos de envío y distribución de categorías inclina la balanza hacia la Tienda 1.

Por lo tanto, se concluye que la Tienda 1 representa la mejor oportunidad comercial para maximizar ventas y satisfacción del cliente para el Sr. Juan.
