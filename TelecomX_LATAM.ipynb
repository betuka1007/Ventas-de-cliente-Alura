#ðŸ”§ TransformaciÃ³n

```python
pd_telecomm_customer = pd.json_normalize(pd_telecomm['customer'])
```

```python
pd_telecomm_phone = pd.json_normalize(pd_telecomm['phone'])
```

```python
pd_telecomm_internet = pd.json_normalize(pd_telecomm['internet'])
```

```python
pd_telecomm_account = pd.json_normalize(pd_telecomm['account'])
```

```python
pd_telecomm_Churn = pd_telecomm['Churn']
```

```python
pd_telecomm = pd.concat([pd_customerID, pd_telecomm_customer, pd_telecomm_phone, pd_telecomm_internet, pd_telecomm_account, pd_telecomm_Churn], axis=1)
```

```python
churn_genre = pd_telecomm[pd_telecomm['Churn'] == 'Yes'].groupby(['gender'])['Churn'].count()
```

```python
churn_genre
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(churn_genre.index, churn_genre.values,
        color=['#1f77b4', '#ff7f0e'])

# Agregar los valores numÃ©ricos encima de cada barra
for index, value in enumerate(churn_genre.values):
    plt.text(index, value + 0.5,  # PosiciÃ³n del texto (x, y)
             str(value),          # Texto a mostrar (convertimos el valor a string)
             ha='center',         # AlineaciÃ³n horizontal centrada
             va='bottom')         # AlineaciÃ³n vertical en la parte inferior

# PersonalizaciÃ³n adicional
plt.xlabel('Gender')
plt.ylabel('Churn Count')
plt.title('Churn Count by Gender')
plt.ylim(0, max(churn_genre.values) * 1.1)  # Ajustar el lÃ­mite superior del eje Y

plt.show()
```
