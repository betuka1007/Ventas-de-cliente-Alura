```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que ya tienes tu DataFrame llamado pd_telecomm

# Crear el boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=pd_telecomm, x='Churn', y='tenure')

# Calcular estadÃ­sticos por grupo
stats = pd_telecomm.groupby('Churn')['tenure'].describe()

# AÃ±adir anotaciones (mediana, Q1, Q3)
for i, churn_status in enumerate(stats.index):
    q1 = stats.loc[churn_status, '25%']
    median = stats.loc[churn_status, '50%']
    q3 = stats.loc[churn_status, '75%']

    # Mostrar texto sobre la caja
    plt.text(i - 0.2, q1, f'Q1: {q1:.1f}', color='blue', fontsize=9)
    plt.text(i - 0.2, median, f'Med: {median:.1f}', color='red', fontsize=9)
    plt.text(i - 0.2, q3, f'Q3: {q3:.1f}', color='brown', fontsize=9)

# Personalizar el grÃ¡fico
plt.title("DistribuciÃ³n de tiempo de contrato por cancelaciÃ³n")
plt.xlabel("Â¿CancelÃ³?")
plt.ylabel("Meses de contrato")
plt.tight_layout()
plt.savefig("grafico.png")
plt.show()
```
