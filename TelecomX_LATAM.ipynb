```python
import pandas as pd
```

```python
import json
```

```python
telecomm_data = '/content/TelecomX_Data.json'
```

```python
pd_telecomm = pd.read_json(telecomm_data)
```

```python
pd_telecomm.head()
```

```python
pd_telecomm_customer = pd.json_normalize(pd_telecomm['customer'])
```

```python
pd_telecomm_phone = pd.json_normalize(pd_telecomm['phone'])
```

```python
pd_telecomm_internet = pd.json_normalize(pd_telecomm['internet'])
```

```python
pd_telecomm_account = pd.json_normalize(pd_telecomm['account'])
```

```python
pd_customerID = pd_telecomm['customerID']
```

```python
pd_telecomm_Chrurn = pd_telecomm['Churn']
```

```python
pd_telecomm = pd.concat([pd_customerID, pd_telecomm_customer, pd_telecomm_phone, pd_telecomm_internet, pd_telecomm_account, pd_telecomm_Chrurn], axis=1)
```

```python
pd_telecomm.head()
```

```python
pd_telecomm.to_csv('pd_telecomm.csv')
```
