# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Visualización de datos en Python/inmigrantes_canada.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
años = list(map(str, range(1980, 2014)))
años
```

```python
colombia = df.loc['Colombia', años]
colombia
```

```python
col_dict = {'Año':colombia.index.to_list(),
            'Inmigrantes':colombia.values.tolist()} # Changed to .tolist()
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Número de inmigrantes')
plt.show()
```
